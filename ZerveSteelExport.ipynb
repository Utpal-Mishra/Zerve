```python
data = data.sort_values(['Year', 'Country'])
data = data[['Year', 'Country', 'Product Trade Value', 'Product Growth', 'Product Growth Rate', 'Gross Trade Value', 'Product Percentage']]
data.head()
```

# **Visualization**

```python
def barplot(data):
  fig = px.bar(data, x=data.Country, y=data['Product Trade Value'], animation_frame=data.Year, color=data.Country)
  fig.update_xaxes(title_text = "Countries", rangeslider_visible=False, showline=True, linewidth=2, linecolor='black', mirror=True)
  fig.update_yaxes(title_text = 'Trade Value', showline=True, linewidth=2, linecolor='black', mirror=True)
  fig.update_layout(height=600, width=1400, title_text='Top 10 Countries Product Trade Values') 
  fig.show()

barplot(data)
```
